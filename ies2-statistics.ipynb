```python
################################################################################

# Importing Math module
import math
# Importing numpy library
import numpy as np

################################################################################

# Function to find mean.
def mean(arr):
  sum = 0
  for a in arr:
    sum += a
  return sum / len(arr)

# Function to find variance.
def variance(arr, mean, gl):
  sum = 0
  for a in arr:
    sum += pow(a - mean, 2)
  return sum / gl

# Function to find standard deviation.
def stddev(variance):
  return math.sqrt(variance)

# Function to find covariance.
def covariance(arr1, arr2, mean1, mean2, n, gl):
  sum = 0
  for i in range(n):
    sum += (arr1[i] - mean1) * (arr2[i] - mean2)
  return sum / gl

# Function to find correlation.
def correlation(arr1, arr2, mean1, mean2, sd1, sd2, n, gl):
  sum = 0
  for i in range(n):
    sum += (arr1[i] - mean1) * (arr2[i] - mean2)
  return sum / (gl * sd1 * sd2)

################################################################################

# Observations
arrs =[
  [0, 3, 4],
  [1, 2, 4],
  [3, 4, 5],
]
# Number of observations
numob = len(arrs)
# Number of variables
numov = len(arrs[0])
# Degrees of freedom
gl = numov - 1
# Means
means = []
# General mean
gmean = 0
# Variances
variances = []
# Standard deviations
stddevs = []
# Covariance
obs_covariance = 0
# Covariance matrix
mcov = []
for i in range(numov):
  aux = []
  for j in range(numov):
    aux.append(0)
  mcov.append(aux)
# Correlation matrix
mcorr = []
for i in range(numov):
  aux = []
  for j in range(numov):
    aux.append(0)
  mcorr.append(aux)

print('Observations:')
for o in arrs:
  print(o)
print()

################################################################################

# Find means
print('Mean of each observation:')
for arr in arrs:
  m = mean(arr)
  means.append(m)
print(means, end='\n\n')

# Find general mean
print('General mean:')
gmean = sum(means) / numob
# Variables:
print(gmean, end='\n\n')

# Find variances
print('Variance of each observation:')
for i, arr in enumerate(arrs):
  v = variance(arr, means[i], gl)
  variances.append(v)
print(variances, end='\n\n')

# Find standard deviation
print('Standard deviation of each observation:')
for variance in variances:
  sd = stddev(variance)
  stddevs.append(sd)
print(stddevs, end='\n\n')

# Find covariance matrix
print('Covariance matrix:')
for i in range(numob):
  for j in range(numob):
    mcov[i][j] = covariance(arrs[i], arrs[j], means[i], means[j], numov, gl)
for i in mcov:
  print(i)
print()

# Find correlation matrix
print('Correlation matrix:')
for i in range(numob):
  for j in range(numob):
    mcorr[i][j] = correlation(arrs[i], arrs[j], means[i], means[j],
      stddevs[i], stddevs[j], numov, gl)
for i in mcorr:
  print(i)
print()

# Find covariance matrix
x = np.array(arrs)
print('Covariance matrix:')
print(np.cov(x), end='\n\n')

# Find correlation matrix
print('Correlation matrix:')
print(np.corrcoef(x), end='\n\n')

################################################################################
```
